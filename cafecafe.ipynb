```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    "Ao": [2019, 2020, 2021, 2022, 2023],
    "Demanda": [2.0, 2.3, 2.3, 2.3, 2.2],
    "Oferta": [1.4, 2.4, 2.3, 1.9, 2.3],
    "Demanda Insatisfecha": [-0.6, 0.1, 0.0, -0.4, 0.1],
}

df = pd.DataFrame(data)

X = df[["Ao"]]  
y_demanda = df["Demanda"]  

model_demanda = LinearRegression()
model_demanda.fit(X, y_demanda)

y_oferta = df["Oferta"] 

model_oferta = LinearRegression()
model_oferta.fit(X, y_oferta)

aos_futuros = np.array([2025, 2026, 2027, 2028]).reshape(-1, 1)
predicciones_demanda = model_demanda.predict(aos_futuros)
predicciones_oferta = model_oferta.predict(aos_futuros)

# Grfico de Demanda
plt.figure(figsize=(10, 6))
plt.scatter(df["Ao"], df["Demanda"], color="blue", label="Datos histricos de Demanda")
plt.plot(np.append(df["Ao"].values, aos_futuros.flatten()), 
         model_demanda.predict(np.append(df["Ao"].values.reshape(-1, 1), aos_futuros, axis=0)), 
         color="blue", linestyle="--", label="Prediccin de Demanda")
plt.scatter(aos_futuros, predicciones_demanda, color="cyan", label="Predicciones futuras de Demanda")
plt.xlabel("Ao")
plt.ylabel("Demanda")
plt.title("Regresin Lineal - Estimacin de la Demanda")
plt.legend()
plt.grid()
plt.show()

# Grfico de Oferta
plt.figure(figsize=(10, 6))
plt.scatter(df["Ao"], df["Oferta"], color="orange", label="Datos histricos de Oferta")
plt.plot(np.append(df["Ao"].values, aos_futuros.flatten()), 
         model_oferta.predict(np.append(df["Ao"].values.reshape(-1, 1), aos_futuros, axis=0)), 
         color="orange", linestyle="--", label="Prediccin de Oferta")
plt.scatter(aos_futuros, predicciones_oferta, color="red", label="Predicciones futuras de Oferta")
plt.xlabel("Ao")
plt.ylabel("Oferta")
plt.title("Regresin Lineal - Estimacin de la Oferta")
plt.legend()
plt.grid()
plt.show()

# Mostrar predicciones para Demanda y Oferta
print("Predicciones para los aos 2025-2028:")
print("\nDemanda:")
for ao, pred in zip(aos_futuros.flatten(), predicciones_demanda):
    print(f"Ao: {ao}, Prediccin de Demanda: {pred:.2f}")

print("\nOferta:")
for ao, pred in zip(aos_futuros.flatten(), predicciones_oferta):
    print(f"Ao: {ao}, Prediccin de Oferta: {pred:.2f}")
```
